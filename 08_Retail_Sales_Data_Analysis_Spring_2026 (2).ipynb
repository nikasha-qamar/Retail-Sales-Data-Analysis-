### Question 4: Monthly Revenue Calculation (10 Marks)

Calculate the total revenue ('Total Amount') for each month. The result should be a new DataFrame or Series indexed by the month. Display the result.

**Hint:** Group the `df_sales` DataFrame by the month of the 'Date' column. You can access the month period using `df_sales['Date'].dt.to_period('M')`. Then, get the sum of 'Total Amount'.

**Expected Output:** A Series with month-year periods as the index and the corresponding total revenue as values.

````python
# Write your answer here
totalamount=(df_sales.groupby(df_sales['Date'].dt.to_period('M'))['Total Amount']).sum()
print(totalamount)
````

### Question 5: Weekly Revenue Calculation (10 Marks)

Similar to the last question, now calculate the total revenue ('Total Amount') for each week. Display the head of the resulting Series.

**Hint:** Group the `df_sales` DataFrame by the week of the 'Date' column. You can use `df_sales['Date'].dt.to_period('W')`. Then, get the sum of 'Total Amount' and display the first few results with `.head()`.

**Expected Output:** The first 5 rows of a Series with week periods as the index and total revenue as values.

````python
# Write your answer

totalamount_week=(
    df_sales.groupby(df_sales['Date'].dt.to_period('W'))['Total Amount'].sum()
    )
print("Weekly Revenue:\n",totalamount_week)
````

### Question 6: Best-Selling Product Category (10 Marks)

Determine which 'Product Category' has the highest total revenue. Group the data by 'Product Category', sum the 'Total Amount', and find the category with the maximum sum.

**Hint:** Use `.groupby('Product Category')['Total Amount'].sum()` and then find the maximum value and its index using `.idxmax()` and `.max()`.

**Expected Output:** A printout of the name of the best-selling category and its total revenue.

````python
# Write your answer here
totalrevenue=(
    df_sales.groupby('Product Category')['Total Amount'].sum()
)
maxtotalrev=totalrevenue.max()
totalrev_category=totalrevenue.idxmax()
print("Best-Selling Category: ",totalrev_category)
print("Total Revenue of Best-Selling Category: Rupees",maxtotalrev)
````

### Question 7: Top 5 Customers by Spending (10 Marks)

Identify the top 5 customers who have spent the most in total. Group the data by 'Customer ID', sum their 'Total Amount', and display the top 5.

**Hint:** Use `.groupby('Customer ID')['Total Amount'].sum()`, then `.sort_values(ascending=False)`, and finally `.head(5)`.

**Expected Output:** A Series showing the Customer IDs of the top 5 spenders and their total spending.

````python
# Write your answer here
top5=(
    df_sales.groupby('Customer ID')['Total Amount'].sum().sort_values(ascending=False)
)
print("Top 5 Customers and their Spending:\n",top5.head(5))
````

### Question 8: Average Order Value per Customer (15 Marks)

Calculate the average order value (AOV) for each customer. Group by 'Customer ID' and find the mean of the 'Total Amount'. Display the AOV for the top 5 customers by average spending.

**Hint:** This is very similar to the previous question, but you will use `.mean()` instead of `.sum()`.

**Expected Output:** A Series showing the Customer IDs of the top 5 customers by average order value and their AOV.

````python
# Write your answer here
top5=(
    df_sales.groupby('Customer ID')['Total Amount'].mean().sort_values(ascending=False)
)
print("Top 5 Customers and their Average Order Values:\n")
top5.head(5)
````

### Question 9: Cumulative Revenue Calculation (15 Marks)

Calculate and display the cumulative sum of revenue over time. First, group the data by 'Date' and sum the 'Total Amount'. Then, calculate the cumulative sum on this result. Display the last 5 days of the cumulative revenue series.

**Hint:** First, create a daily revenue series: `daily_revenue = df_sales.groupby('Date')['Total Amount'].sum()`. Then, use the `.cumsum()` method on `daily_revenue`. Use `.tail()` to see the final values.

**Expected Output:** A Series showing the last 5 dates and the cumulative total revenue up to each of those dates.

````python
# Write your answer here
daily_revenue=(
    df_sales.groupby('Date')['Total Amount'].sum()
)
crevenue=daily_revenue.cumsum()
print("Cumulative Revenue of last 5 days:\n",crevenue.tail())
````

### Question 10: Pivot Table for Category vs. Month (15 Marks)

Create a pivot table to show the total 'Total Amount' for each 'Product Category' aggregated by month. First, extract the month number into a 'Month' column. Then create the pivot table with 'Month' as the index and 'Product Category' as columns. Display the resulting pivot table.

**Hint:** Create the 'Month' column using `df_sales['Date'].dt.month`. Then, use `df_sales.pivot_table(index='Month', columns='Product Category', values='Total Amount', aggfunc='sum')`.

**Expected Output:** A table with months (1-12) as rows, product categories as columns, and the total sales for each intersection.

````python
# Write your answer here
df_sales['Month']=df_sales['Date'].dt.month
table=df_sales.pivot_table(index='Month',columns='Product Category',values='Total Amount',aggfunc='sum')
print("Pivot Table for Category vs. Month:\n",table)
````
